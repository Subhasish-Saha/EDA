```python
#loading the necessary libraries
import pandas as pd
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
fifa = pd.read_csv('/content/drive/MyDrive/DATASETS/players_20.csv') #loading the dataset
fifa
```

```python
#finding the number of players with same nationality
fifa['nationality'].value_counts()
```

```python
fifa['nationality'].value_counts()[0:10]  #top 10 countries with highest players
```

```python
fifa['nationality'].value_counts()[0:10].keys() #name of the top 10 countries with highest number of players
```

```python
#plotting a bar plot depicting the top 5 countries with number of players 
plt.figure(figsize=(5,5))
plt.bar(list(fifa['nationality'].value_counts()[0:5].keys()),list(fifa['nationality'].value_counts()[0:5]),color = "orange")
plt.show()
```

```python
player_salary = fifa[['short_name','wage_eur']] #extracting name of a player and his salary
```

```python
player_salary.head()
```

```python
#sorting salary of players in ascending order
player_salary = player_salary.sort_values(by=['wage_eur'],ascending=False)
```

```python
player_salary.head()
```

```python
#plotting bar plot with respect to the salary of the players
plt.figure(figsize=(8,5))
plt.bar(list(player_salary['short_name'])[0:5],list(player_salary['wage_eur'])[0:5],color = {'blue','green','red','pink','orange'})
plt.show()
```
